```python
def fonction_1(x1, x2):
    return (x1**2 - x2) / x1 - np.exp(x2) 
```

```python
x1 = np.linspace(-3,3,600)
x2 = np.linspace(-3,3,600)
x1_vals, x2_vals = np.meshgrid(x1, x2)
z = fonction_1(x1_vals, x2_vals)
```

```python
fig1 = plt.figure(figsize=(12, 6))
fig1X = fig1.add_subplot(projection='3d')
fig1X.plot_surface(x1_vals, x2_vals, z, cmap='viridis')
fig1X.set_title('Surface de la fonction de Hassane')
fig1X.set_xlabel('x1')
fig1X.set_ylabel('x2')
fig1X.set_zlabel('f(x1, x2)')
plt.show()
```

```python
def tangente1(f, fPrime, a):
    fg = plt.figure(figsize=(12, 6))
    fgc = fg.add_subplot()
    # courbe
    x = np.linspace(-100, 100, 300)
    y = f(x)
    fgc.plot(x, y)
    # tangente
    t = fPrime(a) * (x- a) + f(a)
    fgc.plot(x, t)
    # show
    plt.show()
```

```python
def f1Prime(x1):
    return 2*x1 - 2
```

```python
tangente1(f1, f1Prime, 1)
```

```python
def tangente2(f, fPrime, a):
    fig = plt.figure(figsize=(12, 6))
    fg = fig.add_subplot(projection='3d')
    x1, x2 = np.linspace(-100, 100, 300), np.linspace(-100, 100, 300)
    x1_grid, x2_grid = np.meshgrid(x1, x2)
    # courbe
    z = f(x1_grid, x2_grid)
    fg.plot_surface(x1_grid, x2_grid, z, cmap='viridis')
    # tangente
    grad = fPrime(a[0], a[1])
    t = (
        f(a[0], a[1])+
        grad[0] * (x1_grid - a[0]) +
        grad[1] * (x1_grid - a[1])
    )
    fg.plot_surface(x1_grid, x2_grid, t, cmap='coolwarm')
    # afficher
    plt.show()
```

```python
def rosenbrock(x1, x2):
    return (1 - x1)**2 + 100 * (x2 - x1**2)**2

def gradient_rosenbrock(x1, x2):
    df_dx1 = -2 * (1 - x1) - 400 * x1 * (x2 - x1**2)
    df_dx2 = 200 * (x2 - x1**2)
    return np.array([df_dx1, df_dx2])
```

```python
tangente2(rosenbrock, gradient_rosenbrock, [1,2])
```
